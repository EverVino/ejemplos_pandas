## Por qué Pandas?

Pandas permite una mejor manipulaciones, combinado con python hace que el tratamiento de datos sea muy flexible.

```python
import pandas as pd

datos_covid = pd.read_csv("owid-covid-data.csv")
#df_xlsx = pd.read_excel("owid_data.xlsx")
#print(datos_covid)
print(datos_covid.head(5))
```

```python
datos_covid["location"]
```

```python
datos_covid.iloc[1:5]
```

```python
# Seleccionando columnas
covid_SA = datos_covid[["continent","location", "date","new_cases_smoothed"]]
covid_SA
```

```python
#Filtrando por filas
covid_SA = covid_SA[covid_SA["continent"]=="South America"]
covid_SA
```

```python
covid_SA = covid_SA[covid_SA["date"]>"2021-01-01"]
covid_SA
```

```python
covid_SA = covid_SA.reset_index()
```

```python
covid_SA = covid_SA[["location","date","new_cases_smoothed"]]
covid_SA
```

```python
covid_nSA = covid_SA.pivot(index="date", columns="location", values="new_cases_smoothed")
covid_nSA
```

```python
covid_nSA.plot.line()
```

```python
#Guardando los datos filtrados
covid_nSA.to_csv("covid_south_america.csv")
```

```python
#Describiendo los datos
covid_nSA.describe()
```

```python
# Convertir las columnas a numpy arrays
fechas = covid_nSA.index.values
bolivia = covid_nSA["Bolivia"].values
print(type(fechas))
print(type(bolivia))
#print(len(bolivia))
```

```python
covid_region = covid_nSA[["Bolivia"]]
covid_region.plot.line()
```

```python
# Usando groupby con las funciones sum() mean() count()
covid_cont = datos_covid[["continent", "date", "new_cases_smoothed"]]
covid_cont = covid_cont.groupby(["continent", "date"]).sum().reset_index("continent")
covid_cont
```

```python
covid_ncont = covid_cont.pivot(columns="continent", values="new_cases_smoothed")
covid_ncont
```

```python
covid_ncont.plot.line()
```

```python
# Creando una nueva columna
#covid_indice = datos_covid[["continent", "date", "new_cases_smoothed"]]
print(datos_covid.columns)
```

```python
covid_indice = datos_covid[["continent","date","new_cases_smoothed","new_deaths_smoothed"]]
covid_indice
```

```python
covid_nindice = covid_indice.groupby(["continent", "date"]).sum().reset_index("continent")
covid_nindice["indice"] = covid_nindice["new_deaths_smoothed"]/covid_nindice["new_cases_smoothed"]
covid_nindice
```

```python
covid_nindice = covid_nindice[["continent", "indice"]]
covid_nindice = covid_nindice.pivot(columns="continent", values="indice")
covid_nindice.plot.line()
```
